```python
import pandas as pd
import numpy as np
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
```

```python
r = requests.get('https://www.calendarlabs.com/holidays/brazil/2015')
```

```python
print(r.status_code)
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
holiday = soup.findAll('tr', {'class':["r0", "r1"],})
```

```python
holiday
```

```python
date = [i.find("td", class_="dt_nowrap").text[:12] for i in holiday]
holiday_title = [i.find("a").text for i in holiday]
```

```python
url = []
pages = list(np.arange(2015, 2019,1))

for i in pages: 
    a = f'https://www.calendarlabs.com/holidays/brazil/{i}'
    url.append(a)

url
```

```python
date = []
holiday_title = []

for link in url:
    soup = BeautifulSoup(requests.get(link).text, 'lxml')
    holiday = soup.findAll('tr', {'class':["r0", "r1"],})
    
    for i in holiday:
        date.append(i.find("td", class_="dt_nowrap").text[:12])
        holiday_title.append(i.find("a").text)
```

```python
holiday_df = pd.DataFrame({
    'date': date,
    'holiday_title':holiday_title
})
```

```python
holiday_df.date = pd.to_datetime(holiday_df.date)
```

```python
holiday_df.to_csv('holiday_df.csv', index = False)
```
